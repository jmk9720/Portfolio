```python
import pandas as pd
names =[
        't',
        'q1', 'q2', 'q3',
        'dq1', 'dq2', 'dq3',
        'I1', 'I2', 'I3',
        'eps21', 'eps22', 'eps31', 'eps32',
        'ddq1', 'ddq2', 'ddq3'
        ]
# Time (secs)
# Joint angle
# Joint velocity
# Motor current (A)
# Strain measurements
# Joint accelerations
df_train = pd.read_csv('hw1q5train.csv', header=None,sep=',',
      names=names, index_col=0)
df_train.head(6)
```

```python
import numpy as np
y_train = np.array(df_train[['I2']])
X_train = np.array(df_train[['q2','dq2','eps21', 'eps22', 'eps31', 'eps32','ddq2']])
```

```python
from sklearn import datasets, linear_model
samp, natt = X_train.shape
```

```python
regr = linear_model.LinearRegression()
regr.fit(X_train, y_train)
y_train_pred = regr.predict(X_train)
MSE_train = np.mean((y_train_pred-y_train)**2)/(np.std(y_train)**2)
```

```python
print(MSE_train)
```
